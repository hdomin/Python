# Operaciones con Data Frames


```python
import numpy as np
import pandas as pd
```

Creamos un conjunto de datos imaginario.

A la función DataFrame, se agrega el arreglo "index" que reemplazará los valores correlativos que se generan de forma automática.

```python
index = [ 'Deportivo Suchitepéquez', 'Cremas', 'Rojos', 'Juventud Retalteca']

df = pd.DataFrame( { 
    'Minutos Jugados': [ 90, 80, 100, 95],
    'Goles': [ 4, 2, 1, 2]
}, index = index)

df
```

Si agregamos más índices obtenemos

```python
index.extend(['Deportivo Marquense', 'Petapa'])

df = df.reindex(index)
df
```

En algunas ocaciones es necesario convertir el "index" a una columna.

```python
df = df.reset_index()
df
```

La operación inversa, convertir una columna en el "index" es:

```python
df = df.set_index('index')
df
```

Ahora agregaremos una nueva columna, sus colores

```python
#df = df.drop(['Color'], axis=1)
df.insert( loc=2, column='Color', 
          value=['(Blanco|Azul|Verde)', '(Blanco|Negro)', '(Rojo|Verde)', '(Amarillo|Negro)', '(Amarillo|Negro|Verde)', '(Verde)'] )

df
```

Una función útil es separar las cadenas

```python
df['Color'] = df['Color'].str.strip('(')
df['Color'] = df['Color'].str.strip(')')
df['Color'].str.split('|')
```

Una forma de representar los datos de forma agrupada es utilizando .stack() o. unstack()

```python
df.stack()
```

```python
df.unstack()
```
