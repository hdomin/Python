Para analizar el porcentaje de "missing values" por cada columna, se genera un nuevo data frame y se pondera con el total de registros

```python
x = df.isnull().sum()
y = x / df.shape[0] * 100
z = { "Missing Values": x, "Percent" : y}
ndf = pd.DataFrame(z, columns=["Missing Values", "Percent"])
ndf.sort_values(by="Percent", ascending=False)
```

Al observar los datos, la columna "Loaned From" contiene un alto porcentaje de "missing values" por lo que sería recomendable eliminarla para el análisis si no es posible completarlos.  El comando lleva "axis=1" porque lo que se desea eliminar es la columna.

```python
df = df.drop(['Loaned From'], axis=1)
```
